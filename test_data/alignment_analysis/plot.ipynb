```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
headers = ["incorrect", "correct", "sample_id", "aligner", "country"]
brazildf = pd.read_csv("plot_alns_Brazil.tsv", sep='\t', names=headers, header=None)
nagoyadf = pd.read_csv("plot_alns_Japan.tsv", sep='\t', names=headers, header=None)
nicaraguadf = pd.read_csv("plot_alns_Nicaragua.tsv", sep='\t', names=headers, header=None)
phdf = pd.read_csv("plot_alns_Philippines.tsv", sep='\t', names=headers, header=None)
```

```python
merged = pd.concat([brazildf, nagoyadf, nicaraguadf, phdf])
merged.head()
```

```python
g = sns.FacetGrid(merged, col='country')

g.map(sns.barplot, 'sample_id', 'correct', 'aligner')
g.set_axis_labels('Samples', 'Correctly Aligned')
g.add_legend()
plt.tight_layout()

plt.show()
```

```python
g = sns.FacetGrid(merged, col='country')

g.map(sns.barplot, 'sample_id', 'incorrect', 'aligner')
g.set_axis_labels('Samples', 'Incorrectly Aligned')
g.add_legend()
plt.tight_layout()

plt.show()
```
